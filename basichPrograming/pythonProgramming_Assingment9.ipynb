1.	Write a Python program to check if the given number is a Disarium Number?

```python
n=input("input a number ")
pos=0
sum=0
for i in n:
    pos+=1
    sum+=int(i)**pos
if int(n) == sum:
    print(n,"is a Disarium number")
else:
    print(n," is not a Disarium number")

    
#Other approach without converting into string

n2= int(input("enter a number"))
temp=n2
len=0
while(temp>0):
    len+=1
    q=temp//10
    temp=q
temp=n2
sum=0
for i in range(len,0,-1):
    sum+=(temp%10)**i
    temp=temp//10
if n2 == sum:
    print(n2,"is a Disarium number")
else:
    print(n2," is not a Disarium number")

  

```

2.	Write a Python program to print all disarium numbers between 1 to 100?

```python
def disariumCheck(n):
    temp=n
    len=0
    while(temp>0):
        len+=1
        q=temp//10
        temp=q
    temp=n
    sum=0
    for i in range(len,0,-1):
        sum+=(temp%10)**i
        temp=temp//10
        
    if n == sum:
        print(n,"is a Disarium number")
    else:
        print(n," is not a Disarium number")

for i in range(1,101):
    disariumCheck(i)
        
```

4.	Write a Python program to print all happy numbers between 1 and 100?

```python
for i in range(1,101):
    tot=i
    temp=i
    while (tot not in range(0,10)):
        temp=tot
        tot=0
        
        while (temp>0):
            tot+=(temp%10)**2
            temp//=10
    if tot==1:
        print(i, " is a Happy Number")
    else:
        print(i, " is not a Happy Number")
   
```

5.	Write a Python program to determine whether the given number is a Harshad Number?

```python
n= int(input("enter number"))
tot=0
temp=n
while temp>0:
    d=(temp%10)
    tot=tot+d
    temp//=10
if n%tot==0:
    print(n," is a Harshad Number")
else:
    print(n," is not a Harshad Number")
```
